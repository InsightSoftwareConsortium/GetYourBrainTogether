## Getting Started

As you experiment with itkwidgets you will find that there are a lot of great features in the drawer (which can be toggled open/close with the hamburger menu in the upper left corner). Quick tip: Does your data look like it's low resolution? No worries! In order to handle large datasets itkwidgets streams the data in chunks which means that the resolution improves steadily as the data becomes available to the viewer. There will be a spinning circle next to the image label that indicates this process is still in progress. If the circle has disappeared check the resolution drop-down menu - a lower value means higher resolution.

## Segmentation of thoracic structures from CT series

### Run the next cells to download and view the data inline or [click here to view the data immediately](https://kitware.github.io/itk-vtk-viewer/app/?image=https://data.kitware.com/api/v1/file/6467ea1531eeed39540a9bbe/download/idc_ct_image.mha&labelImage=https://data.kitware.com/api/v1/file/6467ea1631eeed39540a9bc1/download/idc_pred_segthor_image_label.mha) (no download needed) with the itk-vtk-viewer!

```python
ct_image_file = pooch.retrieve('https://data.kitware.com/api/v1/file/6467ea1531eeed39540a9bbe/download',
                           fname='idc_ct_image.mha',
                           known_hash='0b5c2d360eca36910fa05bbd6f83287f6e146b8f8d95bcf3a5060aec1c155253',
                           progressbar=True)
ct_image = itk.imread(ct_image_file)
```

```python
pred_segthor_image_file = pooch.retrieve('https://data.kitware.com/api/v1/file/6467ea1631eeed39540a9bc1/download',
                           fname='idc_pred_segthor_image_label.mha',
                           known_hash='d14a4ac969599a7d75b502acf9287039b4f22b9966281adac41b96f3dee7f53f',
                           progressbar=True)
pred_segthor_image = itk.imread(pred_segthor_image_file)
```

```python
segthor_viewer = view(ct_image, label_image=pred_segthor_image)
```
